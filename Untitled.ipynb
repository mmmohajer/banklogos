```python
import pandas as pd
import bs4
from urllib.request import Request
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup as soup
from slugify import slugify
```

```python
homeUrl = "https://bank-code.net/transit-number"

req = Request(homeUrl, headers = {'User-agent': 'Mozilla/5.0'})

uClient = uReq(req)


page_html = uClient.read()

uClient.close()

page_soup = soup(page_html, "html.parser")

containers = page_soup.tbody.findAll("tr")

col_insName = []
col_insNum = []

for container in containers:
    columns = container.findAll("td")
    insName = columns[0].text.strip()
    insNum = columns[1].text.strip()
    insType = columns[2].text.strip()
    if insType == "Bank":
        col_insName.append(insName)
        col_insNum.append(insNum)
        
newdf = pd.DataFrame({"Institution Name": col_insName,
                      "Institution Number": col_insNum})
newdf.sort_values(by = ["Institution Number"], ascending = True, inplace = True)
newdf.reset_index(inplace = True)
newdf.drop("index", axis = 1, inplace = True)

newdf
```

```python
col_institutionNum = []
col_tNum = []
col_branchName = []
col_address = []
col_city = []
col_province = []
headers = {'User-Agent': 'Mozilla/5.0'}
pagination_num = 100
numberOfIns = -1
```

```python
for institute in col_insName:
    print(f"Institute {institute} has been Started!")
    numberOfIns = numberOfIns + 1
    slug = slugify(institute)
    origUrl = f"{homeUrl}/{slug}"
    urls = []
    
    for i in range(pagination_num):
        if i == 0:
            urls.append(origUrl)
        else:
            query = i * 100
            urls.append(f"{origUrl}/{query}")
    
    j = 0
    has_url = True
    
    for url in urls:
        if has_url:
            req = Request(url, headers = headers)
            uClient = uReq(req)
            page_html = uClient.read()
            uClient.close()
            page_soup = soup(page_html, "html.parser")
            if(page_soup and page_soup.tbody and page_soup.tbody.tr):
                j = j + 1
                containers = page_soup.tbody.findAll("tr")
                
                for container in containers:
                    columns = container.findAll("td")
                    if(len(columns) >= 6):
                        tNum = columns[2].text.strip()
                        branchName = columns[3].text.strip()
                        address = columns[4].text.strip()
                        city = columns[5].text.strip()
                        province = columns[6].text.strip()
                        col_institutionNum.append(col_insNum[numberOfIns])
                        col_tNum.append(tNum)
                        col_branchName.append(branchName)
                        col_address.append(address)
                        col_city.append(city)
                        col_province.append(province)
            
            else:
                has_url = False
                break
                
    print(f"Total number of page for Institute {institute} was {j}!")
        
```
